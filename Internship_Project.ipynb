```python
data_set=data_set.drop(['Employee ID'],axis=1)
```

```python
data_set
```

```python
gander_numerical={"Male":1,"Female":0}
data_set['Gender']=data_set['Gender'].map(gander_numerical)
```

```python
data_set
```

```python
service={'Product':1,'Service':0}
data_set['Company Type']=data_set['Company Type'].map(service)
```

```python
whf={"Yes":1,"No":0}
data_set['WFH Setup Available']=data_set['WFH Setup Available'].map(whf)
```

```python
data_set
```

```python
data_set.isna().sum()
```

```python
data_set.dropna(inplace=True)
```

```python
data_set
```

```python
data_set.isna().sum()
```

```python
inputs=data_set.drop(['Burn Rate'],axis=1)
target=data_set['Burn Rate']
```

```python
inputs
```

```python
target
```

```python
inputs['Date of Joining'] = pd.to_numeric(inputs['Date of Joining'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(inputs,target)
```

```python
prediction=model.predict(inputs)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(target,model.predict(inputs))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(target,model.predict(inputs))
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(5,5))
plt.scatter(x=target,y=prediction)
plt.plot()
z=np.polyfit(target,prediction,1)
p=np.poly1d(z)
plt.plot(target,p(target),color="red")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
```
